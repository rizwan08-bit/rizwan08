```python
# prompt: generate a code for piechart for two columns

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and it's already loaded
# Replace 'column1' and 'column2' with the actual column names from your DataFrame
column1 = "Years of Work Experience"
column2 = "Gender"

# Group the data by the first column and sum the values in the second column
data = df.groupby(column1)[column2].count()

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title(f"Pie Chart of {column2} by {column1}")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "your_dataset.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)

# Define columns for scatter plot
x_column = "Years of Work Experience"  # Replace with your column name (X-axis)
y_column = "Age"  # Replace with your column name (Y-axis)

# Create Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df[x_column], df[y_column], color='blue', alpha=0.5, edgecolors='black')

# Add labels and title
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f"Scatter Plot of {x_column} vs {y_column}")

# Show grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```
